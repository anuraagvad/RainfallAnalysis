```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dailydata=pd.read_csv("/content/monsoon_daily_jun_to_oct.csv").set_index('year')
dailydata.head()
```

```python
fig,ax1=plt.subplots(figsize=(15,2))
sns.heatmap(dailydata.loc[range(1955,1970)],vmin=0,vmax=500,cmap='crest',ax=ax1)
fig,ax2=plt.subplots(figsize=(15,2))
sns.heatmap(dailydata.loc[range(2004,2020)],vmin=0,vmax=500,cmap='crest',ax=ax2)
plt.show()
```

```python
fig,ax1=plt.subplots(figsize=(15,2))
sns.heatmap(dailydata.loc[range(1955,1970)],vmin=0,vmax=500,cmap='flare',ax=ax1)
fig,ax2=plt.subplots(figsize=(15,2))
sns.heatmap(dailydata.loc[range(2004,2020)],vmin=0,vmax=500,cmap='flare',ax=ax2)
plt.show()
```
