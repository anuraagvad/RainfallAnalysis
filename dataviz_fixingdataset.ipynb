```python
# Create a list of integer values that represent every day of an year from 1955 to 2019
days = list(range(1, 24180))

# Iterate over the list and remove the elements that would fall on invalid dates
for i in range(len(days)):
    try:
        # Convert the integer value to a date object
        date = datetime.date(1955, 1, 1) + datetime.timedelta(days[i] - 1)

        # Check if the date is valid
        if date.month != (days[i] - 1) // 31 + 1:
            days[i] = None
    except ValueError:
        days[i] = None

# Remove the None values from the list
days = [day for day in days if day is not None]
```
